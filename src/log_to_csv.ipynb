```python
import numpy as np
import pandas as pd
import os
import re
from datetime import datetime
```

```python
folder = "../data"
```

```python
def separate(logs):
    df = []
    for log in logs:
        #LogFormat "%V %h %l %u %t \"%r\" %>s %O \"%{Referer}i\" \"%{UserAgent}i\""
        a = re.split(r'(.*) (.*) (-*) (-*) \[(.*)\] "(.*)" (\d+) (\d+) "(.*)" "(.*)"', log)[1:-1]
        a.insert(5, "-")
        df.append(a)
        if len(a) != 11:
            print(log)
    return np.array(df)

def initialize(df):
    for i in range(len(df)):
        df[i,4] = pd.to_datetime(df[i,4], format="%d/%b/%Y:%H:%M:%S %z")
        if re.search(r'^[A-Z]+ ', df[i,6]):
            splt = re.split(r'([A-Z]+) (.*)', df[i,6])[1:-1]
            df[i,5] = splt[0]
            df[i,6] = splt[1]
        df[i,7] = int(df[i,7])
        df[i,8] = int(df[i,8])

def create_csv(df, name):
    pd.DataFrame(df, columns=["server_name", "IP", "logname", "authenticate", "date", "petition", "URL", "status", "bytes", "referer", "user-agent"]).to_csv("../data/"+name[:-4]+".csv")
```

```python
#LogFormat "%V %h %l %u %t \"%r\" %>s %O \"%{Referer}i\" \"%{UserAgent}i\""
for name in os.listdir(folder):
    if(".log" in name):
        logs = open(folder+"/"+name).readlines()
        print("Processing -->", name)
        df = separate(logs)
        print("Separated")
        initialize(df)
        print("Initialized")
        create_csv(df, name)
        print("CSV created")
```
